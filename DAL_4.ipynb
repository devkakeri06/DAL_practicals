```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({"Year":[2014,2015,2016,2017,2018],"Sales":[2000,3000,4000,3500,6000]})
plt.plot(df["Year"],df["Sales"])
plt.title("Simple Line plot")
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
share=[20, 12, 11, 4, 3]
companies=['Google', 'Microsoft', 'Amazon', 'IBM', 'Apple']
comp =pd.DataFrame({ "share" : share, "companies" : companies})
ax = comp.plot(y= "share", kind="pie", labels = comp["companies"], autopct= '%1.0f%%', legend=False, title="Market Share")
ax.set(ylabel=' ')
```
